````
import pandas as pd
import numpy as np

# Dataset generation (same seed every time)
np.random.seed(141)
n = 500

categories      = ['Web Development', 'Data Analytics', 'ML/AI', 'Design',
                   'Content Writing', 'Digital Marketing']
cities          = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai', 'Pune', 'Kolkata']
freelancer_lvls = ['Rising Talent', 'Level 1', 'Level 2', 'Top Rated']
statuses        = ['Completed', 'In Progress', 'Cancelled']
months          = ['Jan','Feb','Mar','Apr','May','Jun',
                   'Jul','Aug','Sep','Oct','Nov','Dec']

df_raw = pd.DataFrame({
    'project_id'      : [f'P{str(i).zfill(3)}' for i in range(1, n+1)],
    'category'        : np.random.choice(categories, n, p=[0.25,0.20,0.15,0.15,0.15,0.10]),
    'budget_inr'      : np.random.randint(5000, 150001, n),
    'duration_days'   : np.random.randint(3, 91, n),
    'client_rating'   : np.round(np.random.uniform(2.5, 5.0, n), 1),
    'freelancer_level': np.random.choice(freelancer_lvls, n, p=[0.30,0.30,0.25,0.15]),
    'status'          : np.random.choice(statuses, n, p=[0.65,0.25,0.10]),
    'city'            : np.random.choice(cities, n),
    'platform_fee_pct': np.random.choice([10, 15, 20], n, p=[0.20,0.50,0.30]),
    'month'           : np.random.choice(months, n)
})

# Save CSV — your app.py will load this file
df_raw.to_csv('freelancehub_india.csv', index=False)

print(f'Shape: {df_raw.shape}')
print(f'Columns: {list(df_raw.columns)}')
df_raw.head(3)
````

## Section 2 — Concept Notes

### What is Streamlit?
Streamlit converts a plain Python script into an interactive web app — no HTML, no CSS, no JavaScript.
You write Python; Streamlit handles the browser.

**Why it matters for freelance:** A live URL you can share with a client is worth 10 notebooks. Streamlit Cloud (free tier) deploys in one click from GitHub.

---

### Core Building Blocks

| Element | What it does | Code |
|---|---|---|
| **Page config** | Title, icon, layout | `st.set_page_config(...)` |
| **Title / markdown** | Text on page | `st.title()`, `st.markdown()` |
| **Metric card** | KPI number with delta | `st.metric(label, value, delta)` |
| **Sidebar** | Filter panel on left | `st.sidebar.selectbox(...)` |
| **Selectbox** | Single-choice dropdown | `st.sidebar.selectbox(label, options)` |
| **Multiselect** | Multi-choice dropdown | `st.sidebar.multiselect(label, options, default)` |
| **Slider** | Range selector | `st.slider(label, min, max, value=(lo, hi))` |
| **Dataframe** | Interactive table | `st.dataframe(df, use_container_width=True)` |
| **Columns** | Side-by-side layout | `col1, col2 = st.columns(2)` |
| **Plotly chart** | Rich interactive chart | `st.plotly_chart(fig, use_container_width=True)` |
| **Info box** | Highlighted text block | `st.info("text")` |
| **Cache** | No reload on every rerun | `@st.cache_data` |

---

### How Streamlit Works (Execution Model)

```
User interacts with a widget
        ↓
Streamlit reruns the ENTIRE script top to bottom
        ↓
@st.cache_data stops data load from re-running every time
        ↓
Filtered results update instantly on screen
```

This is why `@st.cache_data` is **not optional** — without it, every filter click reloads from disk.

---

### Streamlit App Structure (Best Practice)

```python
# 1. Imports
# 2. st.set_page_config(...)        <- must be FIRST Streamlit call
# 3. @st.cache_data def load_data()
# 4. df = load_data()
# 5. Sidebar filters
# 6. Apply filter logic
# 7. Metrics row
# 8. Charts
# 9. Insights / NRA box
```

## Section 4 — Answer Key

````
# Complete Day141_app.py — Answer Key

app_code = """
import streamlit as st
import pandas as pd
import plotly.express as px

# T1: Page config — must be FIRST Streamlit call
st.set_page_config(
    page_title='FreelanceHub India',
    page_icon='briefcase',
    layout='wide'
)

st.title('FreelanceHub India - Analytics Dashboard')
st.caption('Month 8 | Day 141 | Deepanshu Garg')

# T2: Load data with caching
@st.cache_data
def load_data():
    return pd.read_csv('freelancehub_india.csv')

df = load_data()

# T3: Sidebar selectbox
st.sidebar.header('Filters')

selected_category = st.sidebar.selectbox(
    'Category',
    options=['All'] + sorted(df['category'].unique().tolist())
)

# T4: Sidebar multiselect + slider
selected_cities = st.sidebar.multiselect(
    'City',
    options=sorted(df['city'].unique().tolist()),
    default=sorted(df['city'].unique().tolist())
)

budget_range = st.sidebar.slider(
    'Budget Range (INR)',
    min_value=int(df['budget_inr'].min()),
    max_value=int(df['budget_inr'].max()),
    value=(int(df['budget_inr'].min()), int(df['budget_inr'].max()))
)

# T5: Apply all 3 filters
filtered = df.copy()
if selected_category != 'All':
    filtered = filtered[filtered['category'] == selected_category]
if selected_cities:
    filtered = filtered[filtered['city'].isin(selected_cities)]
filtered = filtered[
    (filtered['budget_inr'] >= budget_range[0]) &
    (filtered['budget_inr'] <= budget_range[1])
]

st.write(f'Showing **{len(filtered)}** projects')
st.divider()

# T6: 4 metric cards
col1, col2, col3, col4 = st.columns(4)
col1.metric('Total Projects', len(filtered))
col2.metric('Avg Budget', f'Rs {filtered["budget_inr"].mean():,.0f}')
col3.metric('Avg Rating', f'{filtered["client_rating"].mean():.2f}')
col4.metric('Completion Rate', f'{(filtered["status"]=="Completed").mean()*100:.1f}%')

st.divider()

# T7 + BONUS: Two charts side by side
col_left, col_right = st.columns(2)

with col_left:
    st.subheader('Projects by Category')
    cat_counts = filtered['category'].value_counts().reset_index()
    cat_counts.columns = ['Category', 'Count']
    fig_bar = px.bar(
        cat_counts, x='Category', y='Count',
        color='Count', color_continuous_scale='Blues',
        title='Project Count by Category'
    )
    fig_bar.update_layout(showlegend=False, coloraxis_showscale=False)
    st.plotly_chart(fig_bar, use_container_width=True)

with col_right:
    # BONUS: horizontal bar chart
    st.subheader('Avg Budget by Category')
    avg_budget = filtered.groupby('category')['budget_inr'].mean().reset_index()
    avg_budget.columns = ['Category', 'Avg Budget (INR)']
    avg_budget = avg_budget.sort_values('Avg Budget (INR)', ascending=True)
    fig_h = px.bar(
        avg_budget, x='Avg Budget (INR)', y='Category',
        orientation='h', color='Avg Budget (INR)',
        color_continuous_scale='Greens',
        title='Average Budget (INR) by Category'
    )
    fig_h.update_layout(coloraxis_showscale=False)
    st.plotly_chart(fig_h, use_container_width=True)

st.divider()

# T8: NRA insight
top_cat    = filtered.groupby('category')['budget_inr'].mean().idxmax()
top_budget = filtered.groupby('category')['budget_inr'].mean().max()

st.info(
    f'NRA INSIGHT | '
    f'Number: {top_cat} projects average Rs {top_budget:,.0f} - '
    f'Rs {top_budget - filtered["budget_inr"].mean():,.0f} above platform average. '
    f'Reason: Higher technical complexity and smaller qualified freelancer supply '
    f'allow {top_cat} specialists to command premium rates. '
    f'Action: Prioritise {top_cat} bids on Upwork; set minimum proposal at '
    f'Rs {top_budget * 0.80:,.0f} to stay competitive while protecting margin. '
    f'Update Fiverr profile to list {top_cat} as primary skill within 7 days.'
)

# Expandable raw data
with st.expander('View Raw Filtered Data'):
    st.dataframe(filtered, use_container_width=True, height=300)
"""

with open('Day141_app_ANSWER.py', 'w') as f:
    f.write(app_code.strip())

print('Answer key written: Day141_app_ANSWER.py')

# Verification numbers
import pandas as pd
df = pd.read_csv('freelancehub_india.csv')
print(f'Total rows            : {len(df)}')
print(f'Avg budget (all data) : Rs {df["budget_inr"].mean():,.0f}')
print(f'Avg rating            : {df["client_rating"].mean():.2f}')
print(f'Completion rate       : {(df["status"]=="Completed").mean()*100:.1f}%')
print(f'Top category (budget) : {df.groupby("category")["budget_inr"].mean().idxmax()}')
print(f'Top cat avg budget    : Rs {df.groupby("category")["budget_inr"].mean().max():,.0f}')
print('\nCategory project counts:')
print(df['category'].value_counts().to_string())
print('\nCategory avg budgets:')
print(df.groupby('category')['budget_inr'].mean().sort_values(ascending=False).apply(lambda x: f'Rs {x:,.0f}').to_string())
````
